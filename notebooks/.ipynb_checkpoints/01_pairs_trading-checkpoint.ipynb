# Pairs Trading — A Practical Introduction

## What is Pairs Trading?

Pairs trading is a **market-neutral** strategy: you simultaneously go long one stock and short another, profiting from the *relative* movement between them — not from the overall market direction.

The core idea:
1. Find two stocks that historically move together (they are **cointegrated**).
2. Construct a **spread** — a linear combination of their prices — that is stationary (mean-reverting).
3. When the spread deviates far from its historical mean, bet on it reverting:
   - Spread too **low** → buy the underperformer, sell the outperformer
   - Spread too **high** → sell the outperformer, buy the underperformer

Classic example: **Coca-Cola (KO)** and **Pepsi (PEP)**. Both are giant beverage companies exposed to the same economic forces. Their stock prices don't move in lockstep every day, but over time the relationship is stable — temporary divergences tend to close.

## 2. Compute the Spread

We don't just subtract prices — we use **OLS regression** to find the *hedge ratio* β:

$$\log(P_A) = \alpha + \beta \cdot \log(P_B) + \varepsilon$$

The spread is then:

$$\text{spread}_t = \log(P_{A,t}) - \beta \cdot \log(P_{B,t})$$

Why log-prices? They make the spread scale-invariant and linearize multiplicative
price relationships — standard practice in stat arb.

Why OLS? We want the *minimum-variance* linear combination. β tells us how many
units of B to hold per unit of A to make the position as mean-reverting as possible.

````python
beta = compute_hedge_ratio(df['close_a'], df['close_b'])
spread = compute_spread(df['close_a'], df['close_b'], beta)

print(f"Hedge ratio β = {beta:.4f}")
print(f"Spread mean   = {spread.mean():.4f}")
print(f"Spread std    = {spread.std():.4f}")
````

````python
fig = plot_spread(spread, beta)
fig.show()
````

> **What to notice:** A well-behaved spread oscillates around a constant mean without
> drifting upward or downward over time. That's what stationarity looks like visually.

## 3. Statistical Tests

### 3a. ADF Test — Is the spread stationary?

The **Augmented Dickey-Fuller (ADF) test** formally tests whether a time series has
a unit root (non-stationary) or not (stationary).

- **Null hypothesis H₀**: the spread has a unit root (NOT stationary — bad for us)
- **p-value < 0.05**: reject H₀ → spread IS stationary → pairs trading applies ✓
- **p-value ≥ 0.05**: cannot reject H₀ → spread may not be stationary → risky

````python
adf = adf_test(spread)
print(f"ADF statistic : {adf['adf_stat']}")
print(f"p-value       : {adf['p_value']}")
print(f"Stationary?   : {adf['is_stationary']}  {'✓ Good — spread is stationary' if adf['is_stationary'] else '✗ Caution — spread may not be stationary'}")
````

### 3b. Half-Life of Mean Reversion

The **half-life** tells us how many trading days it typically takes for the spread
to revert halfway back to its mean. It is estimated by fitting an AR(1) process:

$$\Delta \text{spread}_t = \alpha \cdot \text{spread}_{t-1} + \varepsilon_t$$

$$\text{half-life} = \frac{-\ln(2)}{\alpha}$$

This is practical guidance:
- **Half-life 10–40 days** → well-suited for swing trading
- **Half-life < 5 days** → more of an intraday/short-term play
- **Half-life > 60 days** → very slow, requires patience and large drawdown tolerance
- A good starting point for your **rolling window** is 2–3× the half-life

````python
hl = compute_half_life(spread)
print(f"Half-life: {hl} trading days")
print(f"Suggested rolling window: {int(hl * 2)}–{int(hl * 3)} days")
````

## 4. Generate Trading Signals

### The Z-Score

Rather than trading off the raw spread level, we use a **rolling z-score**:

$$z_t = \frac{\text{spread}_t - \mu_{t,w}}{\sigma_{t,w}}$$

Where $\mu_{t,w}$ and $\sigma_{t,w}$ are the rolling mean and standard deviation
over the last $w$ days. The z-score is dimensionless — it tells you how many
standard deviations the spread is from its recent average.

### Entry / Exit Rules

| Condition | Action | Meaning |
|---|---|---|
| z < −2 | Long spread | Buy KO, short PEP |
| z > +2 | Short spread | Short KO, buy PEP |
| \|z\| < 0.5 | Exit | Spread reverted to mean |
| \|z\| > 3 | Stop | Spread moved against us |

We only enter when flat — no doubling up on an open position.

````python
params = DEFAULT_PARAMS

zscore = compute_zscore(spread, window=params['rolling_window'])
signals = generate_signals(
    zscore,
    z_entry=params['z_entry'],
    z_exit=params['z_exit'],
    z_stop=params['z_stop'],
)

# Summary of signal events
event_counts = signals['signal'].value_counts().dropna()
print("Signal event counts:")
print(event_counts.to_string())
````

````python
fig = plot_zscore(signals, params['z_entry'], params['z_exit'], params['z_stop'])
fig.show()
````

> **What to notice:**
> - Blue up-triangles = enter long spread (buy KO, short PEP)
> - Orange down-triangles = enter short spread (short KO, buy PEP)
> - Green circles = exits (spread reverted — take profit)
> - Red X's = stops (spread moved against us — cut loss)
>
> Each entry should be followed by an exit or stop. Longer trades mean slower mean reversion.

## 5. Try Another Pair — JPM / BAC

JPMorgan (JPM) and Bank of America (BAC) are both large US banks exposed to the same
interest rate environment, credit cycle, and regulation. A classic cointegrated pair.

````python
df2 = fetch_pair('JPM', 'BAC', period=PERIOD)
beta2 = compute_hedge_ratio(df2['close_a'], df2['close_b'])
spread2 = compute_spread(df2['close_a'], df2['close_b'], beta2)
adf2 = adf_test(spread2)
hl2 = compute_half_life(spread2)

print(f"Hedge ratio β = {beta2:.4f}")
print(f"ADF p-value   = {adf2['p_value']}  {'✓ stationary' if adf2['is_stationary'] else '✗ not stationary'}")
print(f"Half-life     = {hl2} days")
````

````python
zscore2 = compute_zscore(spread2, window=params['rolling_window'])
signals2 = generate_signals(zscore2, z_entry=params['z_entry'], z_exit=params['z_exit'], z_stop=params['z_stop'])

plot_prices(df2, 'JPM', 'BAC').show()
plot_zscore(signals2, params['z_entry'], params['z_exit'], params['z_stop']).show()
````

                                           ↓
